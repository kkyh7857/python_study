## while

```python
# 1~3까지 whlie 문을 사용해서 출력
number = 1
while number <= 3:
    print(number)
    number+=1
```

```python
number = 1 
while True :
    print(number)
    number+=1
    if number == 4 :
        break
```

```python
while True :
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))
    if(num1 == 0 and num2 == 0):
        print('프로그램이 종료되었습니다.')
        break
    print(f'두 정수의 합은 : {num1+num2}')
```

```python
nowkg = int(input('현재 몸무게 : '))
wantkg = int(input('목표 몸무게 : '))
count = 1;
while True :
    nowkg -= int(input(f'{count}주차 감량 몸무게 : '))
    count+=1
    if nowkg < wantkg :
        print(f'{nowkg}kg 달성!! 축하합니다.')
        break
```

## for

```python
hi = '안녕하세요'
for s in hi:
    print(s)
```

```python
list_food = ['햄버거','치킨','피자']
for f in list_food:
    print(f)
```

```python
t1 = (1,2,3,4,5)
for t in t1:
    print(t)
```

```python
score_list = [90,45,70,60,55]
for s in score_list:
    if s>=60 :
        result = '합격'
    else :
        result = '불합격'
    print(f'{score_list.index(s)+1}번 학생은 {result}입니다.')
```

```python
for num in range(1, 11):
    print(num, end=' ')
```

```python
for num in range(97, 76, -1):
    print(num, end=" ")
```

```python
for num in range(23, 40):
    print(num, end=" ")
```

```python
list1 = [[1,2],[3,4],[5,6,7]]
for i in range(len(list1)):
    for j in range(len(list1[i])):
        print(list1[i][j], end=" ")
```

```python
list1 = [[1,2],[3,4],[5,6,7]]
for i in list1:
    print(i, end=" ")
    for j in i:
        print(j, end=" ")
```

```python
num = int(input('정수 입력>>'))
for n in range(1, num+1):
    if num%n==0:
        print(n, end=' ')
```

```python
ㅌ = int(input('정수입력>>'))
print('구구단',ㅌ,'단',sep=' ')
for i in range(1, 10):
    print(ㅌ,'x',i,'=',ㅌ*i)
```

```python
x = int(input('줄 수를 입력하세요'))
for i in range(1, x+1):
    for j in range(i):
        print('*', end = " ")
    print()
```

```python
for i in range(1, 11):
    for j in range(11-i):
        print('*', end= " ")
    print()
```
