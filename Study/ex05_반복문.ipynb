## while

```python
# 1~3 whlie   
number = 1
while number <= 3:
    print(number)
    number+=1
```

```python
number = 1 
while True :
    print(number)
    number+=1
    if number == 4 :
        break
```

```python
while True :
    num1 = int(input('    >> '))
    num2 = int(input('    >> '))
    if(num1 == 0 and num2 == 0):
        print(' .')
        break
    print(f'   : {num1+num2}')
```

```python
nowkg = int(input('  : '))
wantkg = int(input('  : '))
count = 1;
while True :
    nowkg -= int(input(f'{count}   : '))
    count+=1
    if nowkg < wantkg :
        print(f'{nowkg}kg !! .')
        break
```

## for

```python
hi = ''
for s in hi:
    print(s)
```

```python
list_food = ['','','']
for f in list_food:
    print(f)
```

```python
t1 = (1,2,3,4,5)
for t in t1:
    print(t)
```

```python
score_list = [90,45,70,60,55]
for s in score_list:
    if s>=60 :
        result = ''
    else :
        result = ''
    print(f'{score_list.index(s)+1}  {result}.')
```

```python
for num in range(1, 11):
    print(num, end=' ')
```

```python
for num in range(97, 76, -1):
    print(num, end=" ")
```

```python
for num in range(23, 40):
    print(num, end=" ")
```

```python
list1 = [[1,2],[3,4],[5,6,7]]
for i in range(len(list1)):
    for j in range(len(list1[i])):
        print(list1[i][j], end=" ")
```

```python
list1 = [[1,2],[3,4],[5,6,7]]
for i in list1:
    print(i, end=" ")
    for j in i:
        print(j, end=" ")
```

```python
num = int(input(' >>'))
for n in range(1, num+1):
    if num%n==0:
        print(n, end=' ')
```
