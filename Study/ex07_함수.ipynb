```python
def number_sum(n1, n2):
    result = n1 + n2
    return result
```

```python
re = number_sum(7,5)
print(re)
re = number_sum('abc','def')
print(re)
```

```python
def cal(num1, num2, op):
    '''덧셈과 뺄셈을 계산하는 함수'''
    if op == '+':
        return num1+num2
    elif op == '-' :
        return num1-num2
    else :
        return '뭐여! 시방 지금 장난하냐'



num1 = int(input('정수입력 >> '))
num2 = int(input('정수입력 >> '))
op = input('연산자 입력(+,-) >> ')

result = cal(num1, num2, op)

print(f'결과 : {result}')
```

```python
def add(*n):
    return sum(n)
```

```python
def add_sub(num1, num2):
    return num1+num2, num1-num2
```

```python
result = add_sub(4,7)
print(result)
```

```python
re1, re2 = add_sub(4,7)
print(re1, re2)
```

```python
# 매개 변수의 기본값 설정하기
def power_of_N(num, power=2):
    return num**power
```
