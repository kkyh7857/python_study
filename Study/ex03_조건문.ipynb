## 조건문
- 상황에 따라 다르게 처리하는 구문
- 종류: if, elif, else

```python
score = 10
if score >=60 :
    print('합격')
else :
   print('불합격') # 탭 또는 스페이스바 4칸인데 안 맞으면 빨간 색으로 불남
print('입니다')
```

```python
money = int(input('얼마 있어?'))
if money > 10000:
    print('택시를 탄다')
elif money > 5000:
    print('따릉이를 탄다')
elif money > 2000 :
    print('버스를 탄다')
else :
    print('걸어간다')
```

```python
score = int(input('점수입력>>'))
if score >= 90 and score <=100 :
    print('A')
elif score >= 80 :
    print('B')
elif score >= 70 :
    print('C')
elif score >= 60 :
    print('D')
else :
    print('F')
```

```python
score = int(input('점수입력>>'))
if 100 >= score >= 90 :
    point = 'A'
elif 90 > score >= 80 :
    point = 'B'
elif 80 > score >= 70 :
    point = 'C'
elif 70 > score >= 60 :
    point = 'D'
else :
    point = 'F'
print(f'{score}점은 {point}학점 입니다.')
```
