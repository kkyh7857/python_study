### format 함수를 사용한 포매팅
문자열 .format()<Br>
특정 데이터가 들어가야 하는 자리에 {}를 넣어준다.

```python
month = 11
day = 8
tmp = 10

str2 = "오늘의 날짜는 {}월 {}일 이고, 온도는 {}도 입니다.".format(month, day, tmp)
print(str2)
```

### f 문자열을 사용한 포매팅
문자열 가장 앞에 f키워드를 붙여준다.<br>
특정 데이터가 들어가야할 자리에 중괄호를 넣어주고<br>
중괄호 안에 그 데이터도 같이 담아준다.

```python
month = 11
day = 8
tmp = 10

str2 = f"오늘의 날짜는 {month}월 {day}일 이고, 온도는 {tmp}도 입니다."
print(str2)
```

#### 문자열 함수

```python
msg3 = 'Python is easy and simple.'

print(msg3)
# 문자열의 길이 : len(데이터 or 변수)
print(len(msg3)) #length의 줄임말 len()
```

```python
# 양쪽 공백 제거: strip()
msg4 = '         Python is easy and simple.      '
print(msg4)
print(msg4.strip())
```

```python
msg1 = msg4.strip()
msg1
```

```python
# 특정 문자를 다른 문자로 대체 : replace()
# 변수.replace(이전문자, 대체 할 문자)
print(msg1)
print(msg1.replace('Python', 'Java'))
# --> 실제 변수에 저장된 데이터에 영향을 주지 않음
print(msg1)
# --> 대체한 결과를 변수에 반영시키면 된다.
msg1 = msg1.replace('Python', 'Java')
print(msg1)
```
