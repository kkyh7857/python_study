### 배열 연산 함수
- sum, mean

```python
# 2차원 배열 만들기(크기: 2,5)
```

```python
arr2 = np.arange(1,11)
arr2 = arr2.reshape(2,5)
```

```python
# sum
print(f'열 단위 합계 : {np.sum(arr2, axis=0)}') # 열 단위
print(f'행 단위 합계 : {np.sum(arr2, axis = 1)}') # 행 단위
```

```python
# mean
print(f'열 단위 평균: {np.mean(arr2, axis=0)}')
print(f'행 단위 평균: {np.mean(arr2, axis=1)}')
```

```python
data = np.loadtxt('ratings.txt', delimiter='::', dtype='int')
data
```

```python
# 1번 사용자의 데이터만 가져오기
data1 = data[data[:,0]==1] 
data1
# 해석: 첫번째 차원은 전부 다 쓸꺼야, 그런데 그중에 첫번째가 1인 사람만 쓸거야
```

```python
# 1번 사용자의 평점 평균 구하기
np.mean(data1[:,2])
```

```python
# 각 사용자별 평점 평균 구하기
# 결과를 리스트에 저장해서 np배열로 만들기(사용자번호, 평점 평균)
user_list = []
user_num = np.unique(data[:,0])
for i in user_num:
    user1 = data[data[:,0]==i]
    user_avg = np.mean(user1[:,2])
    user_list.append([i, user_avg])
user_list
```

```python
user_avglist = np.array(user_list)
user_avglist
```

```python
# 각 사용자별 평점 평균이 4점 이상인 사용자 구하기
user_avglist[user_avglist[:,1]>=4][:,0].astype(int)
```
