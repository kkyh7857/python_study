```python
# 리스트 요소 추가
list6 = [1,4,8,3,6]
list6.append(10)
list6
# 리스트 제거
list6.remove(1)
list6
# 리스트 삽입
# insert(인덱스 번호, 값)
list6.insert(2, 10)
list6
# 리스트 수정
list6[1] = 3
list6
list6[:] = [1,2,3,4,5,6]# 슬라이싱도 된다.
list6
```

```python
arr = [1,2,3,4,5]
# 5를 7로 변경
arr[4] = 7
arr
# 2를 ['a','b','c']
arr[1] = ['a','b','c']
arr
# 4,7 = 'd','e','f','g'
arr[3:] = 'd','e','f','g'
arr
# 1을 'h','i','j'
arr[:1] = 'h','i','j'
arr
```

```python
# 리스트를 쓰면 for문을 굳이 안써도 될 듯. ....

list = ['딸기', '바나나', '수박', '체리', '포도']
userlist = input('검색할 문자를 입력하세요>>')
if userlist in list :
    print(f'{userlist}는 리스트에 {list.index(userlist)}번째 인덱스에 들어있습니다.')
else :
    print('없슈~ 안 팔아유~')
```
