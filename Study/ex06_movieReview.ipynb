```python
# 텍스트 상태의 데이터가 아닌 파이썬 객체 자체를 저장하고 불러오는 방법
import pickle
```

```python
with open('movie_review.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
count = 0
for i in data:
    if i[1]==1:
        count+=1
print(f"긍정 '{count}'개")
print(f"부정 '{len(data)-count}'개")
```

```python
sum = 0
for i in data:
    sum += len(i[0])
print(f'리뷰의 평균 길이는 "{sum/len(data)}" 입니다.')
```

```python
min = len(data[0][0])
for i in data:
    if len(i[0])< min :
        min = len(i[0])
        review = i[0]
print(f'리뷰 최소 길이는 {min} 입니다.')
print(review)
```

```python
max = len(data[0][0])
for i in data:
    if len(i[0]) > max:
        max = len(i[0])
        review = i[0]
print(f'리뷰 최대 길이는 {max} 입니다.')
print(review)
```

```python
count = 0
for i in data:
    if i[0].count('ㅋ') :
        count+=1
print(f'"ㅋ"는 총 {count} 개 입니다.')
```
