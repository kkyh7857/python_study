#### 연산자
- 종류: 산술연산자, 지수연산자, 대입연산자, 삼항연산자, 비교연산자, 논리연산자
- 나누기 연산자 : /(나누기), //(몫), %(나머지)

```python
num1 = 10
num2 = 7

print(num1 / num2)  # 나누기
print(num1 // num2) # 나누기 몫
print(num1 % num2)  # 나눴을 때 나머지
print(num1 ** num2)
```

```python
# 문자열 더하기
str1 = '안녕'
str2 = '하세요'
print(str1, str2)
print(str1 + str2)
```

```python
str3 = '10'
str4 = '7'
print(str3, str4)
print(str3+str4)
print(10+7)
```

```python
# 숫자 + 문자 : error
num1 = 10
str2 = '7'
print(num1 + str2)
# 서로 다른 데이터타입의 연산을 지원하지 않는다.
# -> 데이터 타입을 맞춰주자
```

```python
# 형변환
# 숫자 + 숫자
print(num1 + int(str2))
# 문자 + 문자
print(str(num1) + str2) 
```

## 키보드로부터 입력받기 : input() 
- 입력한 값을 문자열로 돌려준다!(반환)

```python
num = input('정수를 입력해주세요 : ')
num
# 받은 수는 문자열로 저장 됨
```

```python
# 문자열로 반환 -> 산술연산x
# 해결책 :  문자열을 정수로 변환
num1 = int(input('정수를 입력해주세요 : '))
num1
```

```python
num1 = int(input('정수를 입력하세요>>'))
num2 = int(input('정수를 입력하세요>>'))
print("더하기 결과: ",(num1+num2))
print("빼기 결과: %d" %(num1-num2))
print("곱하기 결과: {}".format(num1*num2))
print(f"나누기 결과: {num1/num2}")
```

```python
# 지수연산자 : 밑수**지수
# 2의 3승은? 8
print(2**3)
print(4**4)
```

```python
# 복합대입연산자
num3 = 27
num3 += 3 # num3 = num3 + 3
print(num3)
num3 -= 3
print(num3)
num3 *= 3
print(num3)
num3 /= 3
print(num3)
```

```python
# 치환
a = 3
b = 20

print(a, b)

temp = a
a = b
b = temp
print(a, b)

# Python 치환
a,b = b,a
print(a,b)
```

```python
# 비교연산자 
# 반환값은 논리값(True/False)
c = 4
d = 15

print(c>d)
print(c<d)
print(c == d)
print(c != d)
```

```python
# not 연산자
print(not True)
print(not False)
print(not 3>20)
```

```python
# and 연산자 - 모두 참이어야 참
print(True and True)
print(False and True)
```

```python
# or 연산자 - 하나라도 참이면 참
print(True or False)
print(False or False)
```

```python
# 삼항 연산자
# a if 조건식 else b
score = 70
print('합격'if score >= 60 else'불합격')
```

```python
# 키보드로 정수를 입력 받아 홀수인지 짝수인지 판별하시오

num6 = int(input('정수 입력>>'))

print(f'{num6}는(은) '+('짝수' if num6%2==0 else '홀수')+' 입니다.' )
```

```python
# 번외1. 문제 정수를 입력 받아 짝수 인지 홀수인지 구별하세요~
num3 = int(input('정수 입력>>'))
if num3%2==0 :
    print('짝수')
else :
    print('홀수')
```

```python
# 번외2.두 수를 입력 받아 모두 짝수면 참 그렇지 않으면 거짓을 출력
num4 = int(input('정수 입력>>'))
num5 = int(input('정수 입력>>'))
print(num4%2==0 and num5%2==0)
```
